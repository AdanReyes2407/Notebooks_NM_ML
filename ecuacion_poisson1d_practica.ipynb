# Poisson 1D: cuaderno guiado con verificaciones

Resolvemos:

$$
-u''(x)=f(x),\quad x\in(0,1),\qquad u(0)=u(1)=0,
$$

con $f(x)=\pi^2\sin(\pi x)$ y solución exacta $u(x)=\sin(\pi x)$.


## Discretización y matriz

En nodos interiores:

$$
\frac{-u_{i+1}+2u_i-u_{i-1}}{h^2}=f_i.
$$

Casos de extremo interior:

$$
\frac{2u_1-u_2}{h^2}=f_1,\qquad
\frac{-u_{N-1}+2u_N}{h^2}=f_N.
$$

La matriz es tridiagonal con diagonal principal $2/h^2$ y sub/superdiagonales $-1/h^2$.


```python
import numpy as np
import matplotlib.pyplot as plt


def exact_solution(x):
    return np.sin(np.pi * x)


def rhs_f(x):
    return (np.pi**2) * np.sin(np.pi * x)
```

## TO DO A: construir sistema lineal


```python
def build_linear_system(n):
    # TO DO
    h = 1.0 / (n + 1)
    x = np.linspace(h, 1.0 - h, n)
    main = (2.0 / h**2) * np.ones(n)
    off = (-1.0 / h**2) * np.ones(n - 1)
    A = np.diag(main) + np.diag(off, 1) + np.diag(off, -1)
    b = rhs_f(x)
    return x, h, A, b
```

## TO DO B: resolver, mostrar matriz y barrer N=1,...,20


```python
def solve_poisson(n):
    x_in, h, A, b = build_linear_system(n)
    u_in = np.linalg.solve(A, b)
    x_full = np.concatenate(([0.0], x_in, [1.0]))
    u_full = np.concatenate(([0.0], u_in, [0.0]))
    u_ex = exact_solution(x_full)
    err_l2 = np.sqrt(np.mean((u_full - u_ex) ** 2))
    return x_full, u_full, u_ex, h, err_l2, A

x, u_num, u_ex, h, err, A = solve_poisson(5)
print("Matriz A para N=5:")
print(A)
print(f"h={h:.6f}, error={err:.3e}")
```

```python
N_vals = np.arange(1, 21)
errs = []
for n in N_vals:
    _, _, _, h, err, _ = solve_poisson(n)
    errs.append(err)

plt.figure(figsize=(8, 4))
plt.plot(N_vals, errs, 'o-', lw=2.0)
plt.title("Error L2 para N=1,...,20")
plt.xlabel("N (nodos interiores)")
plt.ylabel("Error L2")
plt.grid(alpha=0.25)
plt.tight_layout()
plt.show()
```
