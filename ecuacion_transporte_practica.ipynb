# Ecuación de transporte 1D: práctica guiada

Modelo:

$$
\frac{\partial u}{\partial t}+a\frac{\partial u}{\partial x}=0,
\qquad x\in[0,1),\ t\in[0,T].
$$

Condiciones periódicas y perfil inicial no simétrico para seguir el corrimiento.

En este problema, CFL significa **Courant-Friedrichs-Lewy**.


## Esquema upwind para $a>0$

$$
u_i^{n+1}=u_i^n-\nu(u_i^n-u_{i-1}^n),
\qquad
\nu=\frac{a\Delta t}{\Delta x},
\qquad 0\le \nu\le 1.
$$
