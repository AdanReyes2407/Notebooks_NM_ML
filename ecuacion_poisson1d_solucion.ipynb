# Poisson 1D: solución


```python
import numpy as np
import matplotlib.pyplot as plt


def exact_solution(x):
    return np.sin(np.pi * x)


def rhs_f(x):
    return (np.pi**2) * np.sin(np.pi * x)


def build_linear_system(n):
    h = 1.0 / (n + 1)
    x = np.linspace(h, 1.0 - h, n)
    main = (2.0 / h**2) * np.ones(n)
    off = (-1.0 / h**2) * np.ones(n - 1)
    A = np.diag(main) + np.diag(off, 1) + np.diag(off, -1)
    b = rhs_f(x)
    return x, h, A, b


def solve_poisson(n):
    x_in, h, A, b = build_linear_system(n)
    u_in = np.linalg.solve(A, b)
    x_full = np.concatenate(([0.0], x_in, [1.0]))
    u_num = np.concatenate(([0.0], u_in, [0.0]))
    u_ex = exact_solution(x_full)
    err_l2 = np.sqrt(np.mean((u_num - u_ex) ** 2))
    return x_full, u_num, u_ex, h, err_l2, A


N_vals = np.arange(1, 21)
errs = []
for n in N_vals:
    _, _, _, _, err, _ = solve_poisson(n)
    errs.append(err)

print("Error N=1:", errs[0])
print("Error N=20:", errs[-1])
```

```python
plt.figure(figsize=(8, 4))
plt.plot(N_vals, errs, 'o-', lw=2.0)
plt.title("Poisson 1D: error L2 para N=1,...,20")
plt.xlabel("N")
plt.ylabel("Error L2")
plt.grid(alpha=0.25)
plt.tight_layout()
plt.show()
```

## Comparación numérica vs exacta para N=20


```python
x20, u20, u20_ex, _, err20, _ = solve_poisson(20)
print(f"error N=20: {err20:.3e}")
plt.figure(figsize=(8, 4))
plt.plot(x20, u20, lw=2.2, label="Numérica N=20")
plt.plot(x20, u20_ex, "--", lw=2.0, label="Exacta")
plt.title("Poisson 1D: perfil final")
plt.xlabel("x")
plt.ylabel("u(x)")
plt.grid(alpha=0.25)
plt.legend()
plt.tight_layout()
plt.show()
```
