# Boosting Trees

Este método y el método de bosuqe aleatorios son parecdiso en el sentido de que ambos utilizan muchos árboles aletaorios. La diefrencia fundamental entre ellos es la manera en la que se utilizan los datos para constuir dichos árboles. Ambos utilizan estos árboles utilizando particiones aletaorios del set train, estas particiones son de talas iguales y los datso ya seleccionados par netrenar un arbol pueden ser seleccionados para entrenar otro árbol. Esto es lo que llamamos **bagging** o **bootsrap** y es  la manera de utilizar este bootstrap lo que diferencia a ambos métodos. En los bosques aleatorios la probabilidad de tomar un elemento que jamás ha sido tomado para entrenar a un árbol es la misma que tomar un elemento que ya ha sido utilizado anteriormente. En el caso de los **boosting trees** es que la probabilidad de tomar un elemnto que ya fue tomado cambia (mediante un vector de pesos). Estos pesos se eligen en función de que si los datso han sido bien predeciso con los árboles entrenados. Es decir si hay un elemnto que los árboles que ya tenemos entrenado no modelan o predicen bien se le dará mayor prioridad a que lo tomen para entrenamiento los árboles que siguen.
Entonces se usa la información de **accuracy** anterior para intentarla siempre mejorar de forma secuencial.

Hay dos parámetros a especifiacr acá. Éstos son:

- Tree complexity (tc): controla el número de divisiones en cada árbol. 

- Learning rate (lr):Determina la contyribución de cada árbol al modelo genertal. Un valor pequeo significa que hay que utilzar muchos árboles

Generalmente se usan 1000 árboles aproximádamente en total. Para dataset con unos 500 datos o menos los árboles peuqños son recomendables con unos 2 o 3 split (divisiones)

Es una técnica escalable a dataset muy grandes y es robusta a outliers (muy dispersos con respecto a cierta estadística como la media empírica) o valores faltantes.
