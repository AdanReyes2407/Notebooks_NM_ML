# Ecuación de calor 2D en placa: solución


```python
import numpy as np
import matplotlib.pyplot as plt


def apply_bc(u):
    u[:, 0] = 0.0
    u[0, :] = 0.0
    u[:, -1] = 1.0
    u[-1, :] = 1.0
    return u


def build_problem(alpha=1.0, L=1.0, T=0.5, N=34, r_obj=0.22):
    h = L / N
    dt = r_obj * h * h / alpha
    nt = int(np.ceil(T / dt))
    dt = T / nt
    r = alpha * dt / (h * h)
    x = np.linspace(0.0, L, N + 1)
    y = np.linspace(0.0, L, N + 1)
    X, Y = np.meshgrid(x, y, indexing="xy")
    u = 0.25 * np.exp(-((X - 0.35)**2 + (Y - 0.35)**2) / 0.02)
    u = apply_bc(u)
    return x, y, dt, nt, r, u


def step_heat2d(u, r):
    un = u.copy()
    u_new = un.copy()
    u_new[1:-1, 1:-1] = un[1:-1, 1:-1] + r * (
        un[2:, 1:-1] + un[:-2, 1:-1] + un[1:-1, 2:] + un[1:-1, :-2] - 4.0 * un[1:-1, 1:-1]
    )
    return apply_bc(u_new)


def solve_heat2d(alpha=1.0, L=1.0, T=0.5, N=34, r_obj=0.22, store_every=4):
    x, y, dt, nt, r, u = build_problem(alpha=alpha, L=L, T=T, N=N, r_obj=r_obj)
    t_hist = [0.0]
    U_hist = [u.copy()]
    for n in range(nt):
        u = step_heat2d(u, r)
        t_next = (n + 1) * dt
        if (n + 1) % store_every == 0 or (n + 1) == nt:
            t_hist.append(t_next)
            U_hist.append(u.copy())
    return x, y, r, np.array(t_hist), np.array(U_hist)


x, y, r, t_hist, U_hist = solve_heat2d()
print(f"r={r:.6f}")
print(f"t_final={t_hist[-1]:.6f}")
```

```python
plt.figure(figsize=(6.6, 5.3))
im = plt.imshow(U_hist[-1], extent=[0, 1, 0, 1], origin="lower", cmap="turbo", vmin=0.0, vmax=1.0, aspect="equal")
plt.colorbar(im, orientation="horizontal", pad=0.12, label="Temperatura")
plt.title("Calor 2D en t_final")
plt.xlabel("x")
plt.ylabel("y")
plt.tight_layout()
plt.show()
```

## Campos térmicos en instantes de referencia


```python
t_marks = [0.0, 0.16, 0.33, 0.5]
fig, axes = plt.subplots(2, 2, figsize=(9, 7))
for ax, tm in zip(axes.ravel(), t_marks):
    idx = int(np.argmin(np.abs(t_hist - tm)))
    im = ax.imshow(U_hist[idx], extent=[0,1,0,1], origin="lower", cmap="turbo", vmin=0.0, vmax=1.0, aspect="equal")
    ax.set_title(f"t={t_hist[idx]:.3f}")
    ax.set_xlabel("x")
    ax.set_ylabel("y")
cbar = fig.colorbar(im, ax=axes.ravel().tolist(), orientation="horizontal", fraction=0.06, pad=0.08)
cbar.set_label("Temperatura")
fig.tight_layout()
plt.show()
```

## Norma máximo del campo en el tiempo


```python
umax = [float(np.max(np.abs(U_hist[k]))) for k in range(len(t_hist))]
plt.figure(figsize=(7.8, 4))
plt.plot(t_hist, umax, "o-", lw=1.8)
plt.title("Calor 2D: evolución de ||u||_inf")
plt.xlabel("t")
plt.ylabel("||u||_inf")
plt.grid(alpha=0.25)
plt.tight_layout()
plt.show()
```
